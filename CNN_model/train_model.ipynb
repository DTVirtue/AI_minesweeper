```python
import CNN
from Trainer import Trainer
from keras.models import load_model
```

```python
# Create a Trainer object with the CNN model
learner = Trainer(CNN.model)

# Train the model
learner.train_model(
    samples_per_batch=1000,  # Number of samples per batch
    total_batches=10,       # Total number of batches to train
    epochs_per_batch=1     # Number of epochs to train per batch
)

# Save the trained model to a file named "CNN.h5"
learner.model.save("CNN.h5")
```
