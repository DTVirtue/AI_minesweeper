```python
from minesweeper import generate_data
import matplotlib.pyplot as plt
import numpy as np
```

```python
easy_board_size, easy_bomb_count = 9, 10
med_board_size, med_bomb_count = 16, 40
exp_board_size, exp_bomb_count = 22, 99
SP, CSP = 0, 1
RANDOM, LOCAL_PROB = 0, 1
```

```python
data_SP_random_easy = generate_data(easy_board_size, easy_bomb_count, SP, RANDOM, iterations = 1000)
data_SP_local_easy = generate_data(easy_board_size, easy_bomb_count, SP, LOCAL_PROB, iterations = 1000)
```

```python
data_SP_random_med = generate_data(med_board_size, med_bomb_count, SP, RANDOM, iterations = 1000)
data_SP_local_med = generate_data(med_board_size, med_bomb_count, SP, LOCAL_PROB, iterations = 1000)
```

```python
data_SP_random_exp = generate_data(exp_board_size, exp_bomb_count, SP, RANDOM, iterations = 1000)
data_SP_local_exp = generate_data(exp_board_size, exp_bomb_count, SP, LOCAL_PROB, iterations = 1000)
```

```python
cumulative_wins_SP_random_easy = np.cumsum(data_SP_random_easy['wins'])
cumulative_wins_SP_local_easy = np.cumsum(data_SP_local_easy['wins'])
cumulative_wins_SP_random_med = np.cumsum(data_SP_random_med['wins'])
cumulative_wins_SP_local_med = np.cumsum(data_SP_local_med['wins'])
cumulative_wins_SP_random_exp = np.cumsum(data_SP_random_exp['wins'])
cumulative_wins_SP_local_exp = np.cumsum(data_SP_local_exp['wins'])

win_percentage_SP_random_easy = cumulative_wins_SP_random_easy / np.arange(1, len(data_SP_random_easy) + 1)
win_percentage_SP_local_easy = cumulative_wins_SP_local_easy / np.arange(1, len(data_SP_local_easy) + 1)
win_percentage_SP_random_med = cumulative_wins_SP_random_med / np.arange(1, len(data_SP_random_med) + 1)
win_percentage_SP_local_med = cumulative_wins_SP_local_med / np.arange(1, len(data_SP_local_med) + 1)
win_percentage_SP_random_exp = cumulative_wins_SP_random_exp / np.arange(1, len(data_SP_random_exp) + 1)
win_percentage_SP_local_exp = cumulative_wins_SP_local_exp / np.arange(1, len(data_SP_local_exp) + 1)

plt.figure(figsize=(10,6))

plt.plot(win_percentage_SP_local_easy, color='green', label='SP_local_easy')
plt.plot(win_percentage_SP_random_easy, color='blue', label='SP_random_easy')
plt.plot(win_percentage_SP_local_med, color='purple', label='SP_local_med')
plt.plot(win_percentage_SP_random_med, color='red', label='SP_random_med')
plt.plot(win_percentage_SP_local_exp, color='cyan', label='SP_local_exp')
plt.plot(win_percentage_SP_random_exp, color='orange', label='SP_random_exp')

plt.title('SP Solver: Win Percentage Over Time for Different Difficulties')
plt.xlabel('Iterations')
plt.ylabel('Win Percentage')

plt.legend()

plt.grid(True)

plt.show()
```

```python
# Create a function to filter and calculate average times
def get_average_times(data):
    times = data['times'][data['wins'] == 1]
    return times / np.arange(1, len(times) + 1)

# Calculate average times only for wins
average_times_SP_local_easy = get_average_times(data_SP_local_easy)
average_times_SP_random_easy = get_average_times(data_SP_random_easy)
average_times_SP_local_med = get_average_times(data_SP_local_med)
average_times_SP_random_med = get_average_times(data_SP_random_med)
average_times_SP_local_exp = get_average_times(data_SP_local_exp)
average_times_SP_random_exp = get_average_times(data_SP_random_exp)

# Create plot
plt.figure(figsize=(10, 6))

plt.plot(average_times_SP_local_easy, color='green', label='SP_local_easy')
plt.plot(average_times_SP_random_easy, color='blue', label='SP_random_easy')
plt.plot(average_times_SP_local_med, color='purple', label='SP_local_med')
plt.plot(average_times_SP_random_med, color='red', label='SP_random_med')
plt.plot(average_times_SP_local_exp, color='cyan', label='SP_local_exp')
plt.plot(average_times_SP_random_exp, color='orange', label='SP_random_exp')

plt.title('SP Solver: Average Time Over Iterations for Different Difficulties')
plt.xlabel('Iterations')
plt.ylabel('Average Time (s)')

plt.legend()
plt.grid(True)

plt.show()
```

```python
wins_SP_random_easy = data_SP_random_easy['wins']
wins_SP_local_easy = data_SP_local_easy['wins']
wins_SP_random_med = data_SP_random_med['wins']
wins_SP_local_med = data_SP_local_med['wins']
wins_SP_random_exp = data_SP_random_exp['wins']
wins_SP_local_exp = data_SP_local_exp['wins']

# Filter the move_counts based on wins
move_counts_SP_random_easy = data_SP_random_easy['move_counts'][wins_SP_random_easy == 1]
move_counts_SP_local_easy = data_SP_local_easy['move_counts'][wins_SP_local_easy == 1]
move_counts_SP_random_med = data_SP_random_med['move_counts'][wins_SP_random_med == 1]
move_counts_SP_local_med = data_SP_local_med['move_counts'][wins_SP_local_med == 1]
move_counts_SP_random_exp = data_SP_random_exp['move_counts'][wins_SP_random_exp == 1]
move_counts_SP_local_exp = data_SP_local_exp['move_counts'][wins_SP_local_exp == 1]

# Compute average move counts
avg_move_counts_SP_random_easy = np.cumsum(move_counts_SP_random_easy) / np.arange(1, len(move_counts_SP_random_easy) + 1)
avg_move_counts_SP_local_easy = np.cumsum(move_counts_SP_local_easy) / np.arange(1, len(move_counts_SP_local_easy) + 1)
avg_move_counts_SP_random_med = np.cumsum(move_counts_SP_random_med) / np.arange(1, len(move_counts_SP_random_med) + 1)
avg_move_counts_SP_local_med = np.cumsum(move_counts_SP_local_med) / np.arange(1, len(move_counts_SP_local_med) + 1)
avg_move_counts_SP_random_exp = np.cumsum(move_counts_SP_random_exp) / np.arange(1, len(move_counts_SP_random_exp) + 1)
avg_move_counts_SP_local_exp = np.cumsum(move_counts_SP_local_exp) / np.arange(1, len(move_counts_SP_local_exp) + 1)

plt.figure(figsize=(10, 6))

plt.plot(avg_move_counts_SP_local_easy, color='green', label='SP_local_easy')
plt.plot(avg_move_counts_SP_random_easy, color='blue', label='SP_random_easy')
plt.plot(avg_move_counts_SP_local_med, color='purple', label='SP_local_med')
plt.plot(avg_move_counts_SP_random_med, color='red', label='SP_random_med')
plt.plot(avg_move_counts_SP_local_exp, color='cyan', label='SP_local_exp')
plt.plot(avg_move_counts_SP_random_exp, color='orange', label='SP_random_exp')

plt.title('SP Solver: Average Move Count Over Iterations for Different Difficulties (Wins Only)')
plt.xlabel('Iterations')
plt.ylabel('Average Move Count')

plt.legend()
plt.grid(True)

plt.show()
```

```python
# Create a function to filter and calculate average times
def get_average_times(data):
    times = data['times'][data['wins'] == 1]
    # Return NaN or the calculated average times depending on if times array is empty or not
    return times / np.arange(1, len(times) + 1) if times.size > 0 else np.array([np.NaN]*len(data['times']))

# Calculate average times only for wins
average_times_CSP_local_easy = get_average_times(data_CSP_local_easy)
average_times_CSP_random_easy = get_average_times(data_CSP_random_easy)
average_times_CSP_local_med = get_average_times(data_CSP_local_med)
average_times_CSP_random_med = get_average_times(data_CSP_random_med)
average_times_CSP_local_exp = get_average_times(data_CSP_local_exp)
average_times_CSP_random_exp = get_average_times(data_CSP_random_exp)

# Create plot
plt.figure(figsize=(10, 6))

plt.plot(average_times_CSP_local_easy, color='green', label='CSP_local_easy')
plt.plot(average_times_CSP_random_easy, color='blue', label='CSP_random_easy')
plt.plot(average_times_CSP_local_med, color='purple', label='CSP_local_med')
plt.plot(average_times_CSP_random_med, color='red', label='CSP_random_med')
plt.plot(average_times_CSP_local_exp, color='cyan', label='CSP_local_exp')
plt.plot(average_times_CSP_random_exp, color='orange', label='CSP_random_exp')

plt.title('CSP Solver: Average Time Over Iterations for Different Difficulties')
plt.xlabel('Iterations')
plt.ylabel('Average Time (s)')

plt.legend()
plt.grid(True)

plt.show()
```
